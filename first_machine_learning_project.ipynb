# DATA SETİNİ DÜZENLEME

```python
# Kütüphaneleri içe aktarıyoruz
from sklearn.preprocessing import MinMaxScaler

# 'EstimatedSalary', 'Cred_Bal_Sal', 'Bal_sal' ve 'Tenure_Age' sütunlarını sayısal türdeki verilere dönüştürüyoruz
df['EstimatedSalary'] = pd.to_numeric(df['EstimatedSalary'], errors='coerce')
df['Cred_Bal_Sal'] = pd.to_numeric(df['Cred_Bal_Sal'], errors='coerce')
df['Bal_sal'] = pd.to_numeric(df['Bal_sal'], errors='coerce')
df['Tenure_Age'] = pd.to_numeric(df['Tenure_Age'], errors='coerce')

# Min-Max ölçekleyici oluşturuyoruz
scaler = MinMaxScaler()

# Ölçeklenecek sütunları belirliyoruz
columns_to_scale = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts',
                    'EstimatedSalary', 'Age_Tenure_product', 'Mem__no__Products', 
                    'Cred_Bal_Sal', 'Bal_sal', 'Tenure_Age']

# Belirtilen sütunlar üzerinde Min-Max ölçekleme uyguluyoruz
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# NaN değer içeren sütunları veri setinden kaldırıyoruz
df.dropna(axis=1, inplace=True)

# 'Surname' sütununu veri setinden çıkar (eğer mevcutsa)
if 'Surname' in df.columns:
    df.drop(columns=['Surname'], inplace=True)

# TF-IDF sütunlarını veri setinden çıkar (eğer mevcutsa)
tfidf_columns = [col for col in df.columns if 'Surname_tfidf_' in col]
if tfidf_columns:
    df.drop(columns=tfidf_columns, inplace=True)

# One-Hot Encoding Kategorik değişkenleri (örn. 'France', 'Germany', 'Spain', 'Female', 'Male') sayısal formatta temsillerine dönüştürüyoruz.
df_encoded = pd.get_dummies(df, columns=['France', 'Germany', 'Spain', 'Female', 'Male'])
```

# GÖZETİMLİ ÖĞRENME İÇİN TRAİN - TEST SPLİT & HEDEF DEĞİŞKEN TANIMLAMA

```python
from sklearn.model_selection import train_test_split

# Özellikler (X) ve hedef değişken (y) tanımlanır
X = df.drop('Exited', axis=1)
y = df['Exited']

# Eğitim ve test kümelerine bölme
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from xgboost import XGBRegressor, XGBClassifier
from sklearn.model_selection import cross_validate
from sklearn.metrics import r2_score, accuracy_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
```

### REGRESYON MODELLERİ

```python
# Regresyon modelleri
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Elastic Net Regression": ElasticNet(),
    "Decision Tree Regression": DecisionTreeRegressor(),
    "KNN Regression": KNeighborsRegressor(),
    "XGBRegressor": XGBRegressor()
}
```

## MODEL EĞİTME VE PERFORMANS DEĞERLENDİRMESİ

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from sklearn.model_selection import cross_validate

for name, model in models.items():
    # Modeli eğitim verisiyle eğitme
    model.fit(X_train, y_train)

    # Test setleri üzerinde tahminler yapma
    y_pred = model.predict(X_test)

    # Regresyon metrikleri
    test_r2 = r2_score(y_test, y_pred)
    test_mae = mean_absolute_error(y_test, y_pred)
    test_mse = mean_squared_error(y_test, y_pred)
    test_mape = mean_absolute_percentage_error(y_test, y_pred)
    
    print(f"Model: {name.upper()}")
    print(f"Test R-Squared Score: {test_r2:.5f}")
    print(f"Test R-Squared Score (Percentage): {test_r2 * 100:.2f}%")
    print(f"Test Mean Absolute Error: {test_mae:.5f}")
    print(f"Test Mean Squared Error: {test_mse:.5f}")
    print(f"Test Mean Absolute Percentage Error: {test_mape:.5f}")
    
    # Cross-validation
    scores = cross_validate(model, X_train, y_train,
                            scoring=['r2', 'neg_mean_absolute_error', 'neg_mean_squared_error', 'neg_mean_absolute_percentage_error'],
                            cv=10, return_train_score=False)
    
    scores_df = pd.DataFrame(scores, index=range(1, 11))
    
    print("Cross-Validation Scores:")
    print(scores_df[['test_r2', 'test_neg_mean_absolute_error', 'test_neg_mean_squared_error', 'test_neg_mean_absolute_percentage_error']].mean().abs().apply("{:.5f}".format))
    
    print("\n############################################################################\n")
```

# MODEL EĞİTME VE GÖRSELLEŞTİRME

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans


# Denetimsiz öğrenme modelleri
unsupervised_models = {
    "PCA": PCA(n_components=2),        # Veriyi 2D'ye indirgeme
    "KMeans": KMeans(n_clusters=3),    # Kümelendirme için KMeans
    
    
}
```

```python
for name, model in unsupervised_models.items():
    if name == "PCA":
        # PCA ile veriyi 2D'ye indirgeme
        transformed_data = model.fit_transform(data_train)
        plt.figure(figsize=(10, 6))
        plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c='blue', label='Data Points')
        plt.title(f"{name} Visualization")
        plt.xlabel('Component 1')
        plt.ylabel('Component 2')
        plt.legend()
        plt.show()
    else:
        # KMeans ve DBSCAN, veriyi kümelere ayırır
        model.fit(data_train)
        if name == "KMeans":
            labels = model.labels_
            plt.figure(figsize=(10, 6))
            plt.scatter(data_train.iloc[:, 0], data_train.iloc[:, 1], c=labels, cmap='viridis')
            plt.title(f"{name} Clustering")
            plt.colorbar(label='Cluster Label')
            plt.xlabel('Feature 1')
            plt.ylabel('Feature 2')
            plt.show()
       
```

```python
from sklearn.decomposition import PCA

# PCA modelini tanımlayıp veriyi 2D'ye indirgeme
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_train)

# Açıklanan varyans oranını hesaplayın
explained_variance = pca.explained_variance_ratio_
print(f"PCA Açıklanan Varyans Oranı: {explained_variance.sum():.4f}")
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# KMeans modelini fit et
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_train)

# Küme etiketlerini al
labels = kmeans.labels_

# Silhouette Skoru
silhouette_avg = silhouette_score(data_train, labels)
print(f"KMeans Silhouette Skoru: {silhouette_avg:.4f}")

# Davies-Bouldin Skoru
db_index = davies_bouldin_score(data_train, labels)
print(f"KMeans Davies-Bouldin Skoru: {db_index:.4f}")
```

# HİPERPARAMETRE OPTİMİZASYONU

```python
pca = PCA(n_components='mle', svd_solver='full') #PCA'nın bileşen sayısını otomatik olarak seçmesini sağlayabiliriz

from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Veri ölçekleme
scaler = StandardScaler()

# PCA ve KMeans modelini içeren pipeline
pipeline = Pipeline([
    ('scaler', scaler),
    ('pca', PCA()),
    ('kmeans', KMeans())
])

# Grid Search için hiperparametre ızgarası
param_grid = {
    'pca__n_components': [2, 5, 10],
    'kmeans__n_clusters': [2, 3, 4, 5]
}

# GridSearchCV tanımlama
grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=1)
grid_search.fit(data_train)

# En iyi hiperparametreleri yazdırma
print(f"En İyi Parametreler: {grid_search.best_params_}")
print(f"En İyi Skor: {grid_search.best_score_}")
```
